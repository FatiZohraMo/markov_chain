```python
clean_punctuation = re.sub(r"""
               [-,.;@#?!&"=:$]+  # Accept one or more copies of punctuation
               \ *           # plus zero or more copies of a space,
               """,
               " ",          # and replace it with a single space
               clean_text, flags=re.VERBOSE)
```

```python
print(clean_punctuation)
```

```python
clean = ""
for car in clean_punctuation:
  if "a" <= car <= "z" or car == " ":
    clean += car
```

# Markov model of order 1

```python
import numpy as np
```

```python
# character and it's index
indice = {chr(ord('a')+i) : i for i in range(26)}#characters and their indices
indice[' '] = len(indice)
p = np.zeros((27,27))#, dtype = int)
p1 = np.zeros((27,27))
```

```python
print(p)
```

```python
print(indice)
```

```python
print(indice['b'])
```

Calculating frequencies of character "a" followed by every character from "a" to ["z" and "white space"].

```python
for k in range(len(clean)-1):
  i, j = indice[clean[k]], indice[clean[k+1]]
  p[i,j]+=1
```

```python
for k in range(len(clean1)-1):
  i, j = indice[clean1[k]], indice[clean1[k+1]]
  p1[i,j]+=1
```

```python
print(p)#frequencies
```

```python
print(p1)#frequencies from state 1 to state 2, states here are single characters
```

```python
#probabilities of the transition matrix
for k in range (len(p)):
  p[k, :] /= sum(p[k, :])
```

```python
#probabilities of the transition matrix
for k in range (len(p1)):
  p1[k, :] /= sum(p1[k, :])
```

```python
score = nbTentatives = 0
for k in range(len(clean1)-1):
  i, j = indice[clean1[k]], indice[clean1[k+1]]
  value = p[i, j]
  score += value
  nbTentatives += 1
```
